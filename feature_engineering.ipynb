```python
# feature engineering visuals
import numpy as np
import matplotlib.pyplot as plt
from rf_feature_engineer import check_impurity, check_acc

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
impurity_results = check_impurity(sample_count=1000000, n_estimators=100, max_depth=20)
```

```python
def bar_chart(data, title='Feature Importance', ylabel='Impurity Impact', save_path='feature_impurity.png'):
    plt.clf()
    feat_labels, vals = zip(*data)
    fig, ax = plt.subplots()
    plt.gcf().set_size_inches(15, 9)
    x_locs = np.arange(len(feat_labels))
    plt.bar(x_locs, vals)
    plt.xticks(x_locs, feat_labels, rotation=90)
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    plt.tight_layout()
    plt.savefig(save_path)
    plt.show()
```

```python
bar_chart(impurity_results)
```

```python
acc_results = check_acc(sample_count=1000000, n_estimators=50, max_depth=20)
```

```python
bar_chart(acc_results, title='Feature Accuracy', ylabel='Accuracy Impact', save_path='feature_accuracy.png')
```
